```python
# Definir a classe livro.
class livro:

 #Representa um livro com título, autor, gênero e quantidade disponível.
    def __init__(self,titulo,autor,genero,quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = int(quantidade)

    def __str__(self):
        return f"Título: {self.titulo} | Autor: {self.autor} | Gênero: {self.genero} | Quantidade: {self.quantidade}"
```
